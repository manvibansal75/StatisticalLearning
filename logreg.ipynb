```python
c1=0
c2=0
c3=0
a=[]
for c in df1['country_displayable_name'].tolist():
    if c=='the United States' or c=='Canada':
        #df1['continent'] = "USA"
        a.append("USA")
        c1=c1+1
    elif c=='Hong Kong' or c=='Singapore' or c=='Australia':
        #df1['continent'] = "Asia"
        a.append("Asia")
        c2=c2+1
    else: 
        #df1['continent'] = "Europe"
        a.append("Europe")
        c3=c3+1
print(c1,c2,c3)
```

```python
c1=0
c2=0
c3=0
a=[]
conti1=[]
conti2=[]
conti3=[]
for c in df1['country_displayable_name'].tolist():
    if c=='the United States' or c=='Canada':
        #df1['continent'] = "USA"
        a.append("USA")
        conti1.append(int("1"))
        conti2.append(int("0"))
        conti3.append(int("0"))
        c1=c1+1
    elif c=='Hong Kong' or c=='Singapore' or c=='Australia':
        #df1['continent'] = "Asia"
        a.append("Asia")
        conti1.append(int("0"))
        conti2.append(int("1"))
        conti3.append(int("0"))
        c2=c2+1
    else: 
        #df1['continent'] = "Europe"
        a.append("Europe")
        conti1.append(int("0"))
        conti2.append(int("0"))
        conti3.append(int("1"))
        c3=c3+1
print(c1,c2,c3)
```

```python
df1['conti1'] = conti1
df1['conti2'] = conti2
df1['conti3'] = conti3
df1.head(5)
```
